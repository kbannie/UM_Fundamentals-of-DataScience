## Q1

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/hmda_ca_c2.csv')

df.head()
```

## Q2


```python
df['age']=df['age'].str.slice(3,)
df['race']=df['race'].str.slice(3,)
df['ethnicity']=df['ethnicity'].str.slice(3,)
df['sex']=df['sex'].str.slice(3,)
df['debt_to_income_ratio']=df['debt_to_income_ratio'].str.slice(3,)
df['good_credit']=df['good_credit'].str.slice(3,)
```

```python
df.head()
```

## Q3

```python
np.percentile(df['income'], [10,70,99])
```

```python

df['income_class']=df['income'].apply(lambda x: 'One Percent' if x>=1216.15 else 'Upper Class' if x>=177 else 'Middle Class' if x>=61 else "Low Income")
```

```python
df.head()
```

## Q4.

```python
dummies=pd.get_dummies(df['ethnicity'], dtype='int')

df=pd.concat([df, dummies], axis=1)
```

```python
df.head()
```

```python
df.drop('ethnicity', axis=1, inplace=True)
```

```python
df.columns
```

```python
df.rename(columns={'Hispanic or Latino':'ethnicity_Hispanic or Latino', 'Not Hispanic or Latino':'ethnicity_Not Hispanic or Latino'}, inplace=True)
```
