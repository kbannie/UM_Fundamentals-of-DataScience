### 1. An overview of Sentiment Analysis Approaches

https://direct.mit.edu/coli/article/42/3/595/1534/Sentiment-Analysis-Mining-Opinions-Sentiments-and

The document “An Overview of Sentiment Analysis Approaches” provides a comprehensive review of the methods used in sentiment analysis. It highlights three primary approaches:

	1.	Machine Learning-based Approach
   This approach trains models using annotated datasets to identify the sentiment polarity (positive, negative, or neutral) of the text. Traditional methods include supervised learning algorithms like Naive Bayes and Support Vector Machines. These models require labeled training data, and accuracy depends heavily on feature selection and preprocessing techniques.

	2.	Lexicon-based Approach
  This approach relies on sentiment lexicons, which are pre-compiled dictionaries of words that have been labeled as positive, negative, or neutral. Lexicon-based methods can perform well in domain-specific contexts, especially when domain-specific lexicons are used. However, they often struggle with complex linguistic structures such as negation.
  
	3.	Hybrid Approach
  This method combines elements of both machine learning and lexicon-based techniques. The hybrid approach aims to improve the accuracy of sentiment analysis by leveraging the strengths of both approaches. For example, a combination of lexicon-based classification with machine learning classifiers has been shown to outperform either method used in isolation.

In terms of sentiment analysis levels, the document identifies four granularity levels:

	•	Word Level: Analyzes individual words, typically adjectives and other parts of speech.
	•	Sentence Level: Classifies sentiment for each sentence, identifying the polarity (positive, negative, or neutral) expressed.
	•	Document Level: Provides an overall sentiment classification for the entire document.
	•	Aspect Level: Identifies sentiment related to specific features or entities within the text (e.g., battery life and camera quality of a smartphone in product reviews).

These approaches and levels are widely applied in fields such as product reviews, social media analysis, and customer feedback systems.

### 2. Sentiment analysis using product review data

https://link.springer.com/article/10.1186/s40537-015-0015-2

The paper “Sentiment Analysis Using Product Review Data” (Fang & Zhan, 2015) delves into sentiment analysis by categorizing sentiment polarity in online product reviews from Amazon. The primary focus is on developing a method to identify positive and negative sentiments using natural language processing (NLP) techniques.

Key Points:

	1.	Data Collection
  The study uses over 5.1 million Amazon reviews across various categories like beauty, books, electronics, and home products.

	2.	Sentiment Polarity Categorization
  
  It proposes a three-phase process:
	•	Negation Phrases Identification: Identifying negative expressions (e.g., “not bad”).
	•	Sentiment Score Computation: Using a mathematical approach to calculate sentiment scores based on the presence of positive/negative words in reviews.
	•	Feature Vector Formation: Creating feature vectors (binary strings) to represent sentiment tokens (words/phrases).

	3.	Classification Models
  
  The researchers compared three machine learning models : Naive Bayes, Support Vector Machines (SVM), and Random Forest.
	•	Random Forest outperformed the others, showing the best results in both sentence-level and review-level categorization tasks.

	4.	Challenges & Future Work
  
  The study acknowledges the difficulty of identifying “implicit sentiments” (neutral words with positive/negative meanings) and proposes future work on expanding feature sets and improving classification for nuanced sentiments.


Evaluation:

The paper presents a robust approach to sentiment analysis, particularly in handling large datasets. The novelty lies in its hybrid method combining phrase detection, sentiment scoring, and machine learning models. However, the study also highlights limitations in detecting subtle sentiments, which is a common challenge in opinion mining. The experiment’s design using product review data and comparing different classifiers is thorough, contributing valuable insights to the field of sentiment analysis.

### 3. Artificial Intelligence and Sentiment Analysis : A Review in Competitive Research

https://www.mdpi.com/2073-431X/12/2/37

The paper “Artificial Intelligence and Sentiment Analysis: A Review in Competitive Research” by Hamed Taherdoost and Mitra Madanchian (2023) reviews how artificial intelligence (AI) enhances sentiment analysis, particularly in competitive business research. The review highlights AI-powered sentiment analysis as a key tool for understanding customer feedback and competitive positioning, using both lexicon-based and machine learning-based approaches. The study explores how businesses can use sentiment analysis to assess market trends, customer satisfaction, and competitor strengths/weaknesses.

Key points include:

	1.	AI in Sentiment Analysis
  AI enhances traditional sentiment analysis by improving the accuracy of detecting emotions in text. By using natural language processing (NLP) and machine learning techniques, AI-based systems can analyze a wide range of unstructured text data, such as customer reviews, social media posts, and product feedback, to determine whether the sentiment is positive, negative, or neutral. The strength of AI lies in its ability to capture complex linguistic nuances, such as sarcasm, negation, and context, which traditional lexicon-based methods may miss.

	2.	Application in Competitive Research
  Sentiment analysis has become a vital part of competitive research. Companies can use AI to analyze customer feedback not only for their own products but also for those of competitors. This provides insights into market perceptions, customer satisfaction levels, and competitive advantages. The ability to compare how different brands are perceived allows businesses to refine their strategies and make more informed decisions.

	3.	Challenges in AI-Based Sentiment Analysis
  Despite the advancements, challenges remain. These include the handling of domain-specific language, managing noise in social media data, and improving the accuracy of sentiment classification for complex emotions. The review suggests that future improvements in hybrid models, which combine machine learning with deep learning techniques, may address some of these limitations.

	4.	Future Directions
  The paper outlines future trends in AI-driven sentiment analysis, such as using more sophisticated algorithms to understand deeper emotional insights, developing personalized AI systems for customer interactions, and applying sentiment analysis to real-time business decision-making.

In conclusion, AI-powered sentiment analysis is a growing area with significant potential for improving competitive research, offering businesses the ability to understand customer needs and market trends with greater precision.

### 3. Proprocessing

```python
df.rename(columns={'Review':'text'}, inplace=True)
df.rename(columns={'Sentiment':'sentiment'}, inplace=True)
df.head()
```

```python
def cleaning(text):
    # converting to lowercase, special characters, punctuations
    text = text.lower() # converting to lowercase
    text = re.sub(r"\b\d+\b", "", text) # removing number
    text = re.sub('<.*?>+', '', text) # removing special characters,
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # punctuations
    text = re.sub('\n', '', text)
    text = re.sub('[’“”…]', '', text)


   # removing short form:

    text=re.sub("isn't",'is not',text)
    text=re.sub("he's",'he is',text)
    text=re.sub("wasn't",'was not',text)
    text=re.sub("there's",'there is',text)
    text=re.sub("couldn't",'could not',text)
    text=re.sub("won't",'will not',text)
    text=re.sub("they're",'they are',text)
    text=re.sub("she's",'she is',text)
    text=re.sub("There's",'there is',text)
    text=re.sub("wouldn't",'would not',text)
    text=re.sub("didn't",'did not',text)
    text=re.sub("haven't",'have not',text)
    text=re.sub("That's",'That is',text)
    text=re.sub("you've",'you have',text)
    text=re.sub("He's",'He is',text)
    text=re.sub("what's",'what is',text)
    text=re.sub("weren't",'were not',text)
    text=re.sub("we're",'we are',text)
    text=re.sub("hasn't",'has not',text)
    text=re.sub("you'd",'you would',text)
    text=re.sub("shouldn't",'should not',text)
    text=re.sub("let's",'let us',text)
    text=re.sub("they've",'they have',text)
    text=re.sub("You'll",'You will',text)
    text=re.sub("i'm",'i am',text)
    text=re.sub("we've",'we have',text)
    text=re.sub("it's",'it is',text)
    text=re.sub("don't",'do not',text)
    text=re.sub("that´s",'that is',text)
    text=re.sub("I´m",'I am',text)
    text=re.sub("it’s",'it is',text)
    text=re.sub("she´s",'she is',text)
    text=re.sub("he’s'",'he is',text)
    text=re.sub('I’m','I am',text)
    text=re.sub('I’d','I did',text)
    text=re.sub("he’s'",'he is',text)
    text=re.sub('there’s','there is',text)

    return text

dt = df['text'].apply(cleaning)
```

```python
dt=pd.DataFrame(dt)
dt['sentiment']=df['sentiment']
dt
```

```python
dt['no_sw'] = dt['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
dt
```

```python
#Working with the most Frequent Words:
from collections import Counter
cnt = Counter()
for text in dt["no_sw"].values:
    for word in text.split():
        cnt[word] += 1
cnt.most_common(10)
temp = pd.DataFrame(cnt.most_common(10))
temp.columns=['word', 'count']
temp
```

```python
px.bar(temp, x="count", y="word", title='Commmon Words in Text', orientation='h',
             width=700, height=700)
```

```python
# Remove the most frequent words:
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])
dt["wo_stopfreq"] = dt["no_sw"].apply(lambda text: remove_freqwords(text))
dt.head()
```

```python
# Lemmatization: Lemmatization is converting the word to its base form or lemma by removing affixes from the inflected words.
# It helps to create better features for machine learning and NLP models hence it is an important preprocessing step.


wordnet_lem = WordNetLemmatizer()

dt['wo_stopfreq_lem'] = dt['wo_stopfreq'].apply(wordnet_lem.lemmatize)
dt
```

### 5. Train-test Split

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
text_counts = cv.fit_transform(nb['review'])
```

```python
from sklearn.model_selection import train_test_split
X=text_counts
y=nb['sentiment']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=30)
```

### 6. Navies Bayes Modelling

```python
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import classification_report, confusion_matrix
CNB = ComplementNB()
CNB.fit(X_train, y_train)

from sklearn import metrics
predicted = CNB.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, y_test)

print('ComplementNB model accuracy is',str('{:04.2f}'.format(accuracy_score*100))+'%')
print('------------------------------------------------')
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))
print('------------------------------------------------')
print('Classification Report:')
print(classification_report(y_test, predicted))
```

```python
from sklearn.naive_bayes import MultinomialNB

MNB = MultinomialNB()
MNB.fit(X_train, y_train)

predicted = MNB.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, y_test)

print('MultinominalNB model accuracy is',str('{:04.2f}'.format(accuracy_score*100))+'%')
print('------------------------------------------------')
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))
print('------------------------------------------------')
print('Classification Report:')
print(classification_report(y_test, predicted))
```

```python
from sklearn.naive_bayes import BernoulliNB

BNB = BernoulliNB()
BNB.fit(X_train, y_train)

predicted = BNB.predict(X_test)
accuracy_score_bnb = metrics.accuracy_score(predicted,y_test)

print('BernoulliNB model accuracy = ' + str('{:4.2f}'.format(accuracy_score_bnb*100))+'%')
print('------------------------------------------------')
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))
print('------------------------------------------------')
print('Classification Report:')
print(classification_report(y_test, predicted))
```

```python
from sklearn.metrics import roc_curve

# predict probabilities for CNB, MNB, BNB models:
CNB_prob = CNB.predict_proba(X_test)
MNB_prob = MNB.predict_proba(X_test)
BNB_prob = BNB.predict_proba(X_test)

# roc curve for models
fpr1, tpr1, thresh1 = roc_curve(y_test, CNB_prob[:,1], pos_label=1)
fpr2, tpr2, thresh2 = roc_curve(y_test, MNB_prob[:,1], pos_label=1)
fpr3, tpr3, thresh3 = roc_curve(y_test, BNB_prob[:,1], pos_label=1)

# roc curve for tpr = fpr
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)


# auc scores
from sklearn.metrics import roc_auc_score
auc_CNB = roc_auc_score(y_test, CNB_prob[:,1])
auc_MNB = roc_auc_score(y_test, MNB_prob[:,1])
auc_BNB = roc_auc_score(y_test, BNB_prob[:,1])
print(auc_CNB, auc_MNB,auc_BNB)

# plot roc curves
plt.plot(fpr1, tpr1, linestyle='--',color='red', label='CNB Model')
plt.plot(fpr2, tpr2, linestyle='--',color='green', label='MNB Model')
plt.plot(fpr3, tpr3, linestyle='--',color='blue', label='BNB Model')
plt.plot(p_fpr, p_tpr, linestyle='--', color='pink')

# title
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')

plt.legend(loc='best')
plt.savefig('ROC',dpi=300)
plt.show();
```
